```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
random.seed(1)
```

```python
dataset = pd.read_csv('diabetes_012_health_indicators_BRFSS2015.csv')
dataset.shape
```

```python
dataset.describe().T
```

```python
dataset.groupby(["Diabetes_012"]).size()
```

```python
# combine the prediabetes and the diabetes into a single class
dataset['Diabetes_012'] = dataset['Diabetes_012'].replace({0:0, 1:1, 2:1})
dataset.groupby(["Diabetes_012"]).size()
```

```python
# Seperate the no diabetes (0) from the diabetes (1) and perform random selection in the majority class
df0 = dataset[dataset['Diabetes_012'] == 0]
df1 = dataset[dataset['Diabetes_012'] == 1]

df0_sample = df0.sample(n=len(df1), random_state=42)

balanced_df = pd.concat([df0_sample, df1])
```

```python
# Shuffle the Dataset using random permutation so the classes are mixed
balanced_df = balanced_df.sample(frac=1, random_state=42)

balanced_df.groupby(["Diabetes_012"]).size()

balanced_df.to_csv('diabetes_012_health_indicators_BRFSS2015_balanced.csv', sep=',', index=False)
```
